```python
flow = torch.load("flow_model.pt")
flow.eval()
```

```python
n_samples = int(1e4)
n_sample_size = int(1e3)

steps = int(n_samples/n_sample_size)

generated = None
for i in range(steps):
    with torch.no_grad():
        new_data = flow.sample(n_sample_size).cpu().numpy()
        if generated is None:
            generated = new_data
        else:
            generated = np.vstack((generated, new_data))
```

```python
np.savetxt('generated_samples.csv', generated, delimiter=',')
```
